# 鐵達尼號 隨機森林

```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
test_df = pd.read_csv("test.csv", encoding="utf-8")
train_df
```

```python
train_df.isna().sum() #統計遺失值
```

```python
test_df.isna().sum()
```

```python
med = train_df.median() #算出所有數值型資料的中位數
med
```

```python
# 填數字
med = train_df.median()
train_df = train_df.fillna(med)
train_df.isna().sum()
```

```python
# 不要重新算趨勢 
test_df = test_df.fillna(med)
test_df.isna().sum()
```

```python
most = train_df["Embarked"].value_counts().idxmax()
most
```

```python
# 文字類型: 填入最常出現的分類
most = train_df["Embarked"].value_counts().idxmax()
train_df["Embarked"] = train_df["Embarked"].fillna(most)
train_df.isna().sum()
```

```python
test_df["Embarked"] = test_df["Embarked"].fillna(most)
test_df.isna().sum()
```

```python
#one hot encoding
dummy = pd.get_dummies(train_df["Embarked"])
train_df = pd.concat([train_df, dummy], axis=1)
dummy = pd.get_dummies(test_df["Embarked"])
test_df = pd.concat([test_df, dummy], axis=1)
```

```python
dummy = pd.get_dummies(train_df["Sex"])
train_df = pd.concat([train_df, dummy], axis=1)
dummy = pd.get_dummies(test_df["Sex"])
test_df = pd.concat([test_df, dummy], axis=1)
```

```python
train_df
```

```python
# Demo:
s = "Braund, Mr. Owen Harris"
s.split(",")[-1].split(".")[0].replace(" ", "")
```

```python
def middle(s):
    return s.split(",")[-1].split(".")[0].replace(" ", "")
train_df["Name"].apply(middle).value_counts()
```

```python
def middle(s):
    mid = s.split(",")[-1].split(".")[0].replace(" ", "")
    reserved = ["Mr", "Mrs", "Miss", "Master"]
    if mid in reserved:
        return mid
    else:
        return "X"
mid = train_df["Name"].apply(middle)
dummy = pd.get_dummies(mid)
train_df = pd.concat([train_df, dummy], axis=1)
mid = test_df["Name"].apply(middle)
dummy = pd.get_dummies(mid)
test_df = pd.concat([test_df, dummy], axis=1)
```

```python
test_df
```

```python
train = train_df.drop(["PassengerId", "Name",
                       "Sex", "Ticket", "Cabin",
                       "Embarked", "X"], axis=1)
y_train = train["Survived"]
x_train = train.drop(["Survived"], axis=1)
x_train
```

```python
test = test_df.drop(["Name", "Sex", 
                     "Ticket", "Cabin",
                     "Embarked", "X"], axis=1)
testid = test["PassengerId"]
x_test = test.drop(["PassengerId"], axis=1)
x_test
```

```python
# https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.figure(figsize=(14, 14))
sns.heatmap(train.corr(),
            annot=True,
            cmap="RdBu_r")
```

```python
# 使用GridSearchCV套件計算最佳的參數，幾個弱分類器 + 深度
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
clf = RandomForestClassifier()
p = {
    "n_estimators":range(20, 50),
    "max_depth":range(5, 20)
}
grid = GridSearchCV(clf, p, cv=10, n_jobs=4)
grid.fit(x_train, y_train)
grid.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
clf = RandomForestClassifier(n_estimators=25, max_depth=9)
s = cross_val_score(clf, x_train, y_train, cv=10) #交叉驗證
print("十次:", s)
print("平均:", np.average(s))
```

```python
#製作模型且輸出結果
clf = RandomForestClassifier(n_estimators=25, max_depth=9)
clf.fit(x_train, y_train)
answer = pd.DataFrame()
answer["PassengerId"] = testid
answer["Survived"] = clf.predict(x_test)
answer.to_csv("titanic_rf.csv", encoding="utf-8", index=False)
answer
```

使用決策樹

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
clf = DecisionTreeClassifier()
p = {
    "max_depth":range(5, 15)
}
grid = GridSearchCV(clf, p, cv=10, n_jobs=4)
grid.fit(x_train, y_train)
grid.best_params_
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
clf = DecisionTreeClassifier(max_depth=6)
s = cross_val_score(clf, x_train, y_train, cv=10) #交叉驗證
print("十次:", s)
print("平均:", np.average(s))
```

```python
#製作模型且輸出結果
clf = DecisionTreeClassifier(max_depth=6)
clf.fit(x_train, y_train)
answer = pd.DataFrame()
answer["PassengerId"] = testid
answer["Survived"] = clf.predict(x_test)
answer.to_csv("titanic_DecisionTreeClassifier.csv", encoding="utf-8", index=False)
answer
```

# KNN

```python
f = pd.DataFrame()
f["name"] = x_train.columns
f["importance"] = clf.feature_importances_
f
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_scale = scaler.fit_transform(x_train)
x_test_scale = scaler.transform(x_test)
pd.DataFrame(x_train_scale)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
clf = KNeighborsClassifier()
p = {
    "n_neighbors":range(5, 100),
}
grid = GridSearchCV(clf, p, cv=10, n_jobs=4)
grid.fit(x_train_scale, y_train)
grid.best_params_
```

```python
clf = KNeighborsClassifier(n_neighbors=17)
clf.fit(x_train_scale, y_train)
answer = pd.DataFrame()
answer["PassengerId"] = testid
answer["Survived"] = clf.predict(x_test_scale)
answer.to_csv("titanic_knn.csv", encoding="utf-8", index=False)
answer
```
