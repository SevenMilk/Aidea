# use feature_selector & auto-sklearn

```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
train_df
```

```python
train_df["1暴露岩岸"] = train_df["1暴露岩岸"].fillna('0')
train_df["2暴露人造結構物"] = train_df["2暴露人造結構物"].fillna('0')
train_df["3暴露岩盤"] = train_df["3暴露岩盤"].fillna('0')
train_df["4沙灘"] = train_df["4沙灘"].fillna('0')
train_df["5砂礫混合灘"] = train_df["5砂礫混合灘"].fillna('0')
train_df["6礫石灘"] = train_df["6礫石灘"].fillna('0')
train_df["7開闊潮間帶"] = train_df["7開闊潮間帶"].fillna('0')
train_df["8遮蔽岩岸"] = train_df["8遮蔽岩岸"].fillna('0')
train_df["9遮蔽潮間帶"] = train_df["9遮蔽潮間帶"].fillna('0')
train_df["10遮蔽濕地"] = train_df["10遮蔽濕地"].fillna('0')
```

```python
dummy = pd.get_dummies(train_df["County"])
train_df = pd.concat([train_df, dummy], axis=1)
dummy = pd.get_dummies(train_df["Location"])
train_df = pd.concat([train_df, dummy], axis=1)
train_df
```

```python
from feature_selector import FeatureSelector
train_labels = train_df.LEVEL # 獲取all features 
train_features = train_df.drop(columns='LEVEL') # 創建 feature-selector 實例，並傳入features 和labels 
fs = FeatureSelector(data=train_features, labels=train_labels)
fs
```

```python
fs.identify_all(selection_params = {'missing_threshold': 0.6,    
                                    'correlation_threshold': 0.75, 
                                    'task': 'classification',    
                                    'eval_metric': 'auc', 
                                    'cumulative_importance': 0.8})
```

```python
train_removed_all = fs.remove(methods = 'all', keep_one_hot=True)
```

```python
train_removed_all
```

```python
train_removed_all = train_removed_all.drop(["Station", "Season", "County", "Location", "縣市"], axis=1)
```

```python
import autosklearn.classification
import sklearn.model_selection
import sklearn.datasets
import sklearn.metrics
x_train, x_vail, y_train, y_vail = sklearn.model_selection.train_test_split(train_removed_all, train_labels, random_state=1)
automl = autosklearn.classification.AutoSklearnClassifier()
automl.fit(x_train, y_train)
y_hat = automl.predict(x_vail)
print("Accuracy score", sklearn.metrics.accuracy_score(y_vail, y_hat))
```

```python
automl.show_models()
```

```python
import pandas as pd
train_df = pd.read_csv("train.csv", encoding="utf-8")
train_df
```

```python
train_df["1暴露岩岸"] = train_df["1暴露岩岸"].fillna('0')
train_df["2暴露人造結構物"] = train_df["2暴露人造結構物"].fillna('0')
train_df["3暴露岩盤"] = train_df["3暴露岩盤"].fillna('0')
train_df["4沙灘"] = train_df["4沙灘"].fillna('0')
train_df["5砂礫混合灘"] = train_df["5砂礫混合灘"].fillna('0')
train_df["6礫石灘"] = train_df["6礫石灘"].fillna('0')
train_df["7開闊潮間帶"] = train_df["7開闊潮間帶"].fillna('0')
train_df["8遮蔽岩岸"] = train_df["8遮蔽岩岸"].fillna('0')
train_df["9遮蔽潮間帶"] = train_df["9遮蔽潮間帶"].fillna('0')
train_df["10遮蔽濕地"] = train_df["10遮蔽濕地"].fillna('0')
```

```python
pd.set_option('max_columns', 50)
train_df
```

```python
#one hot encoding
# dummy = pd.get_dummies(train_df["County"])
# train_df = pd.concat([train_df, dummy], axis=1)
# dummy = pd.get_dummies(train_df["Location"])
# train_df = pd.concat([train_df, dummy], axis=1)
# #dummy = pd.get_dummies(train_df["Station"])
# #train_df = pd.concat([train_df, dummy], axis=1)
# pd.set_option('max_columns', 1000)
# train_df
```

```python
train_df = train_df.drop(["Location", "County", "Station"], axis=1)
train_df
```

```python
from sklearn.model_selection import train_test_split
y_train = train_df['LEVEL']
x_train = train_df.drop(['LEVEL'], axis=1)

x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.3, random_state=1)
```

Auto_Sklearn

```python
import autosklearn.classification
import sklearn.model_selection
import sklearn.datasets
import sklearn.metrics
automl = autosklearn.classification.AutoSklearnClassifier()
automl.fit(x_train, y_train)
y_hat = automl.predict(x_valid)
print("Accuracy score", sklearn.metrics.accuracy_score(y_valid, y_hat))
```

```python
automl.show_models()
```

AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
bdt = AdaBoostClassifier(DecisionTreeClassifier(max_depth=10, min_samples_split=20, min_samples_leaf=5), algorithm="SAMME",n_estimators=200, learning_rate=0.8)
bdt.fit(x_train, y_train)
```

```python
bdt.feature_importances_
```

```python
bdt.predict(x_valid)
```

```python
bdt.score(x_valid,y_valid)
```

Random Forest

```python
# 使用GridSearchCV套件計算最佳的參數，幾個弱分類器 + 深度
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
clf = RandomForestClassifier()
p = {
    "n_estimators":range(50, 80),
    "max_depth":range(10,20)
}
grid = GridSearchCV(clf, p, cv=10, n_jobs=4)
grid.fit(x_train, y_train)
grid.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
clf = RandomForestClassifier(n_estimators=50, max_depth=10)
s = cross_val_score(clf, x_train, y_train, cv=10) #交叉驗證
print("十次:", s)
print("平均:", np.average(s))
```

```python
import sklearn.metrics
clf = RandomForestClassifier(n_estimators=28, max_depth=11)
clf.fit(x_train, y_train)
y_hat = clf.predict(x_valid)
sklearn.metrics.accuracy_score(y_valid, y_hat, normalize=True, sample_weight=None)
```
