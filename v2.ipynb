```python
from sklearn.feature_selection import chi2

# f_classif: ANOVA F-value between label/feature for classification tasks.
# f_regression: F-value between label/feature for regression tasks.
# 返回的列表表示x的每个离散变量与y离散变量卡方检验得到的卡方检验统计量和P值，第二个特征值P值较大，可以去除这个特征
y_train = train_df['LEVEL']
x_train = train_df.drop(['LEVEL'], axis=1)
chi2(x_train,y_train)
```

```python
from sklearn.model_selection import train_test_split
y_train = train_df['LEVEL']
x_train = train_df.drop(['LEVEL'], axis=1)

x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.3, random_state=1)
```

```python
import autosklearn.classification
```

AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
bdt = AdaBoostClassifier(DecisionTreeClassifier(max_depth=10, min_samples_split=20, min_samples_leaf=5), algorithm="SAMME",n_estimators=200, learning_rate=0.8)
bdt.fit(x_train, y_train)
```

```python
bdt.feature_importances_
```

```python
bdt.predict(x_valid)
```

```python
bdt.score(x_valid,y_valid)
```

Random Forest

```python
# 使用GridSearchCV套件計算最佳的參數，幾個弱分類器 + 深度
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
clf = RandomForestClassifier()
p = {
    "n_estimators":range(50, 80),
    "max_depth":range(10,20)
}
grid = GridSearchCV(clf, p, cv=10, n_jobs=4)
grid.fit(x_train, y_train)
grid.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
clf = RandomForestClassifier(n_estimators=50, max_depth=10)
s = cross_val_score(clf, x_train, y_train, cv=10) #交叉驗證
print("十次:", s)
print("平均:", np.average(s))
```

```python
import sklearn.metrics
clf = RandomForestClassifier(n_estimators=28, max_depth=11)
clf.fit(x_train, y_train)
y_hat = clf.predict(x_valid)
sklearn.metrics.accuracy_score(y_valid, y_hat, normalize=True, sample_weight=None)
```
